```python
import pandas as pd
import warnings
import os
import sys
import matplotlib.pyplot as plt
import seaborn as sns

# Get the root directory path
root_dir = os.path.abspath(os.path.join(os.getcwd(), '..'))

# Add the root directory to the path
sys.path.insert(0, root_dir)

# Now you can import config.py from the root directory
from config import DATA_PATH
```

```python
df = pd.read_csv(os.path.join(DATA_PATH, 'RTA Dataset.csv'))
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.set(style='whitegrid')
sns.countplot(x='Day_of_week', hue='Accident_severity', data=df, palette='viridis')
plt.title('Accident Severity by Day of Week', fontsize=18)
plt.xlabel('Day of Week', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.legend(title='Accident Severity', title_fontsize=12, fontsize=10)
plt.show()
```
